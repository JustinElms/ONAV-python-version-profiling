```python
import glob

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import holoviews as hv
from holoviews import opts
from holoviews import dim
hv.extension('bokeh')
```

```python
files_310 = glob.glob("3-10_results/*.csv")
files_312 = glob.glob("3-12_results/*.csv")

results = pd.DataFrame()
```

```python
for file in files_310:
    result = pd.read_csv(file)
    result.drop(columns=["URL", "Git Hash", 'Start Time', "Profile File Path"], inplace=True)
    result["Version"] = ["3.10.5"] * len(result)
    if len(results) == 0:
        results = result
    else:
        results = pd.concat([results, result])

for file in files_312:
    result = pd.read_csv(file)
    result.drop(columns=["URL", "Git Hash", 'Start Time', "Profile File Path"], inplace=True)
    result["Version"] = ["3.12.0"] * len(result)
    if len(results) == 0:
        results = result
    else:
        results = pd.concat([results, result])    

results = results[['Version', 'Test', 'Dataset', 'Variable', 'Response Time (s)']]    
results.columns = ['Version', 'Test', 'Dataset', 'Variable', 'ResponseTime']
```

```python
results
```

```python
mask = results["Variable"]=="psubsurfacechannel"
results = results[~mask]
mask = results["Test"]=="observation"
results = results[~mask]
mask = results["Test"]=="class4"
results = results[~mask]

results.to_csv("results.csv")
results
```
